```python
import json
import glob
import pandas as pd
```

```python
RESULTS_PATH = "./hps_search_results/"
```

```python
results = [] 
for path in glob.glob(f"{RESULTS_PATH}/8c654681-6872-4346-9289-40b836a3aaf7/*"):
    with open(path, "r") as f:
        results.append(json.load(f))
        
results = pd.DataFrame(results)
```

```python
results["last_val_acc"] = results["val_accuracy"].apply(lambda val_accs: val_accs[-1])
```

```python
results.sort_values("last_val_acc", ascending=False)
```
